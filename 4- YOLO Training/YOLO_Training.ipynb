```python
from ultralytics import YOLO
from ultralytics.engine.results import Results
from ultralytics.utils.metrics import DetMetrics
```

```python
# Load the pre-trained YOLOv11 model
model: YOLO = YOLO("yolo11n.pt")

# Train the model on the COCO dataset
# URL: https://github.com/ultralytics/ultralytics/blob/main/ultralytics/cfg/datasets/coco128.yaml
results: dict | None = model.train(data="coco128.yaml", epochs=100, imgsz=640)
```

```python
for metric, value in results.results_dict.items():
    print(f'{metric}: {value}'.replace('metrics/', ''))

model.save()
```

```python
# Validate the model
metrics: DetMetrics = model.val(imgsz=640, batch=16)

print('\nAverage Precision across multiple IoU thresholds from 0.5 to 0.95')
print(metrics.box.map)

print('\nAverage Precision at IoU threshold 0.5')
print(metrics.box.map50)

print('\nAverage Precision at IoU threshold 0.75')
print(metrics.box.map75)

print('\nAverage Precision across multiple IoU thresholds from 0.5 to 0.95 of each category')
print(metrics.box.maps)
```
